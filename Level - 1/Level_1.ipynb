# Importing Libraries

```python
# Import core libraries
import pandas as pd  # For data manipulation and analysis
import numpy as np   # For numerical operations

# Import visualization libraries
import matplotlib.pyplot as plt  # For creating plots
import seaborn as sns  # For enhanced data visualization
```

# Load Dataset

```python
# Load the dataset
dataset_path = r"C:\Users\DELL\OneDrive\Desktop1\Data Analyst Tasks - Cognifyz Technologies\Level - 1\Dataset .csv"
data = pd.read_csv(dataset_path)

# Display the first few rows
data.head()
```

# Check for Missing Values

```python
# Check for missing values
missing_values = data.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

# Dataset Overview

```python
# Overview of the dataset
data.info()
```

# Check for Duplicate Rows

```python
# Check for duplicate rows
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

# Basic Statistical Summary

```python
# Statistical summary of the dataset
data.describe()
```

# Determine the top three most common cuisines

```python
# Splitting cuisines and counting the occurrences of each cuisine
cuisine_counts = data['Cuisines'].str.split(', ').explode().value_counts()

# Display the top three most common cuisines
top_cuisines = cuisine_counts.head(3)
top_cuisines
```

# Calculate the percentage of restaurants serving each of the top cuisines

```python
# Total number of restaurants
total_restaurants = len(data)

# Calculate the percentage for each top cuisine
top_cuisine_percentages = (top_cuisines / total_restaurants) * 100
top_cuisine_percentages
```

# Identify the City with the Highest Number of Restaurants

```python
# Count the number of restaurants in each city
city_restaurant_counts = data['City'].value_counts()

# Identify the city with the highest number of restaurants
top_city = city_restaurant_counts.idxmax()
top_city_count = city_restaurant_counts.max()

# Display the result
print(f"The city with the highest number of restaurants is {top_city} with {top_city_count} restaurants.")
```

Visualization

```python
# Plot the top 10 cities with the highest number of restaurants
plt.figure(figsize=(12, 6))
city_restaurant_counts.head(10).plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cities with the Highest Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Calculate the Average Rating for Restaurants in Each City

```python
# Calculate the average rating for each city
average_rating_by_city = data.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)

# Display the top 10 cities with their average ratings
average_rating_by_city.head(10)
```

Visualization

```python
# Plot the top 10 cities with the highest average ratings
plt.figure(figsize=(12, 6))
average_rating_by_city.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities with the Highest Average Ratings')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Determine the City with the Highest Average Rating

```python
# Identify the city with the highest average rating
top_rated_city = average_rating_by_city.idxmax()
highest_average_rating = average_rating_by_city.max()

# Display the result
print(f"The city with the highest average rating is {top_rated_city} with an average rating of {highest_average_rating:.2f}.")
```

# Visualize the Distribution of Price Ranges

```python
# Count the number of restaurants in each price range
price_range_counts = data['Price range'].value_counts().sort_index()

# Create a bar chart to visualize the distribution
plt.figure(figsize=(8, 6))
price_range_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range (1: Lowest, 4: Highest)')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Calculate the Percentage of Restaurants in Each Price Range

```python
# Calculate the percentage of restaurants in each price range
total_restaurants = len(data)
price_range_percentages = (price_range_counts / total_restaurants) * 100

# Display the percentages in a readable format
price_range_percentages_df = price_range_percentages.reset_index()
price_range_percentages_df.columns = ['Price Range', 'Percentage']

# Display the data
price_range_percentages_df
```

# Pie Chart for Price Range Percentages

```python
# Create a pie chart for price range percentages
plt.figure(figsize=(8, 8))
price_range_percentages.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']
)
plt.title('Percentage Distribution of Price Ranges')
plt.ylabel('')  # Remove y-label for better aesthetics
plt.show()
```

# Determine the Percentage of Restaurants that Offer Online Delivery

```python
# Count restaurants offering and not offering online delivery
online_delivery_counts = data['Has Online delivery'].value_counts()

# Calculate the percentage of restaurants offering online delivery
online_delivery_percentages = (online_delivery_counts / len(data)) * 100

# Display the results
print("Percentage of Restaurants Offering Online Delivery:")
print(online_delivery_percentages)
```

# Compare the Average Ratings of Restaurants With and Without Online Delivery

```python
# Calculate the average ratings for restaurants with and without online delivery
average_ratings_online_delivery = data.groupby('Has Online delivery')['Aggregate rating'].mean()

# Display the results
print("Average Ratings of Restaurants Based on Online Delivery:")
print(average_ratings_online_delivery)
```

# Pie Chart for Online Delivery Percentage

```python
# Create a pie chart for online delivery percentages
plt.figure(figsize=(8, 8))
online_delivery_percentages.plot(
    kind='pie',
    autopct='%1.1f%%',
    labels=['No', 'Yes'],
    startangle=140,
    colors=['#FF9999', '#66B2FF']
)
plt.title('Percentage of Restaurants Offering Online Delivery')
plt.ylabel('')  # Remove y-label for better aesthetics
plt.show()
```

# Bar Chart for Average Ratings Comparison

```python
# Create a bar chart for average ratings
plt.figure(figsize=(8, 6))
average_ratings_online_delivery.plot(kind='bar', color=['skyblue', 'orange'], edgecolor='black')
plt.title('Comparison of Average Ratings: Online Delivery vs. No Online Delivery')
plt.xlabel('Online Delivery')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
